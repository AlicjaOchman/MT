## Models to be optimized: Random Forest, XGBoost

```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
dataset = pd.read_csv("/home/alicja/Documents/ultimateData.csv")
dataset.set_index('label')
```

```python
labels = dataset['value'].values
features = dataset.drop('value', axis=1).values
```

```python
features = features[:,1:].astype(np.float32)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.33, random_state=42)
```

```python
from imblearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.metrics import roc_auc_score
from xgboost import XGBClassifier
```

```python
%%time
# A parameter grid for XGBoost
params = {
        'min_child_weight': [1, 5, 10],
        'gamma': [0.1, 0.5, 1, 1.5, 2, 5, 10],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [3, 4, 5, 6, 7, 8],
        'learning_rate': [0.001, 0.01, 0.1, 0.3]
        }
```

```python
xgb = XGBClassifier(n_estimators=600, objective='binary:logistic', silent=True, nthread=1)
```

```python
%%time
random_search = RandomizedSearchCV(xgb, param_distributions=params, n_iter=200, scoring='roc_auc', n_jobs=4, cv=5, verbose=3, random_state=1001 )
```

```python
%%time
random_search.fit(X_train, y_train)
```
